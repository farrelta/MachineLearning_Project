**Feature Engineering**

```python
df_train.duplicated().sum()
```

```python
df_train.drop(['id', 'EmployeeCount', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'MonthlyRate', 'Over18', 'JobSatisfaction', 'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsSinceLastPromotion', 'YearsWithCurrManager', 'JobInvolvement', 'JobLevel'], axis=1, inplace=True)
df_test.drop(['EmployeeCount', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'MonthlyRate', 'Over18', 'JobSatisfaction', 'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsSinceLastPromotion', 'YearsWithCurrManager', 'JobInvolvement', 'JobLevel'], axis=1, inplace=True)
```

```python
df_train = pd.get_dummies(df_train,columns=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'])
df_test = pd.get_dummies(df_test,columns=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'])
```

```python
# Example for a binary column 'Gender' with values 'Male' and 'Female'
df_train['Gender_encoded'] = df_train['Gender'].map({'Male': 1, 'Female': 0})
df_test['Gender_encoded'] = df_test['Gender'].map({'Male': 1, 'Female': 0})

# Example for a binary column 'OverTime' with values 'Yes' and 'No'
df_train['OverTime_encoded'] = df_train['OverTime'].map({'Yes': 1, 'No': 0})
df_test['OverTime_encoded'] = df_test['OverTime'].map({'Yes': 1, 'No': 0})

df_train = df_train.select_dtypes(exclude=['object'])
df_test = df_test.select_dtypes(exclude=['object'])
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Step 4: Initialize the Iterative Imputer
imputer = IterativeImputer(max_iter=10, random_state=0)

# Step 5: Perform MICE Imputation
df_train_imputed = imputer.fit_transform(df_train)
df_test_imputed = imputer.fit_transform(df_test)

# Convert the imputed data back to a DataFrame
df_train = pd.DataFrame(df_train_imputed, columns=df_train.columns)
df_test = pd.DataFrame(df_test_imputed, columns=df_test.columns)
```

```python
y_train = df_train['Attrition'].values
X_train = df_train.drop(['Attrition'], axis=1).values
X_test  = df_test.drop(['id'], axis=1).values
```

```python
# Convert both int64 and float64 columns to float32
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
```

```python
# Revert to SMOTE for class imbalance handling
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# Feature Selection based on importance (RandomForest)
selector = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42))
selector.fit(X_train, y_train)
X_train = selector.transform(X_train)
X_test = selector.transform(X_test)

# Expand polynomial features to include more relevant numeric columns
important_num_cols = ['MonthlyIncome', 'TotalWorkingYears', 'Age', 'DistanceFromHome']  # Include more features
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)

X_train_poly = pd.DataFrame(poly.fit_transform(X_train[:, :len(important_num_cols)]),  
                            columns=poly.get_feature_names_out(important_num_cols))

X_test_poly = pd.DataFrame(poly.transform(X_test[:, :len(important_num_cols)]),
                           columns=poly.get_feature_names_out(important_num_cols))

# Concatenate polynomial features with the original data
X_train = np.concatenate([X_train, X_train_poly], axis=1)
X_test = np.concatenate([X_test, X_test_poly], axis=1)
```

```python
# Normalize Data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Methods**

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import RandomizedSearchCV, KFold
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
import numpy as np
import pandas as pd
import warnings

# Function to evaluate the model
def evaluate_model(model, X_val, y_val):
    y_val_pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, y_val_pred)
    f1 = f1_score(y_val, y_val_pred)
    precision = precision_score(y_val, y_val_pred)
    recall = recall_score(y_val, y_val_pred)
    roc_auc = roc_auc_score(y_val, model.predict_proba(X_val)[:, 1])  # Probability of the positive class
    return accuracy, f1, precision, recall, roc_auc

# Function for Randomized Search and Cross Validation
def randomized_search_model(model, param_distributions, X, y, k=5):
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_distributions,
        scoring='f1',
        cv=kf,
        verbose=2,
        n_jobs=-1,
        random_state=42,
        n_iter=10  # Increased iterations for better hyperparameter tuning
    )
    
    try:
        random_search.fit(X, y)
    except Exception as e:
        print(f"Error during fitting model: {e}")
        return None, None, None
    
    best_model = random_search.best_estimator_
    return best_model, random_search.best_params_, random_search.best_score_

# Define parameter ranges for each model dynamically
def get_param_distributions(model_name):
    if model_name == 'Random Forest':
        return {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5],
            'class_weight': ['balanced']
        }
    elif model_name == 'XGBoost':
        return {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 6, 10]
        }
    elif model_name == 'LightGBM':
        return {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [10, 20],
            'class_weight': ['balanced']
        }
    elif model_name == 'CatBoost':
        return {
            'iterations': [100, 200],
            'depth': [4, 6],
            'learning_rate': [0.01, 0.1],
            'l2_leaf_reg': [1, 3]
        }
    elif model_name == 'Gradient Boosting':
        return {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 4, 5],
            'min_samples_split': [2, 5]
        }

# List of models
model_dict = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, random_state=42),
    'LightGBM': LGBMClassifier(random_state=42),
    'CatBoost': CatBoostClassifier(verbose=0, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(class_weight='balanced', max_iter=500, random_state=42)
}

# Lists to store accuracy, F1 score, precision, recall, ROC AUC, best parameters, and best model
models = ['Random Forest', 'XGBoost', 'LightGBM', 'CatBoost', 'Gradient Boosting', 'Logistic Regression']
results = []

# Loop through each model, perform Randomized Search, and evaluate
for model_name in models:
    model = model_dict[model_name]
    param_distributions = get_param_distributions(model_name)
    
    best_model, best_params_model, best_score = randomized_search_model(model, param_distributions, X_train, y_train, k=5)
    
    if best_model is not None:
        accuracy, f1, precision, recall, roc_auc = evaluate_model(best_model, X_train, y_train)
        results.append({
            'Model': model_name,
            'Accuracy': accuracy,
            'F1 Score': f1,
            'Precision': precision,
            'Recall': recall,
            'ROC AUC': roc_auc,
            'Best Params': best_params_model
        })

# Create DataFrame to display results
results_df = pd.DataFrame(results)

# Automatically select the best model based on the highest F1 score
if not results_df.empty:
    best_model_row = results_df.loc[results_df['F1 Score'].idxmax()]

    print(f"The best model based on F1 score is: {best_model_row['Model']}")
    print(f"Best parameters for {best_model_row['Model']}: {best_model_row['Best Params']}")
else:
    print("No models were successfully fitted.")

# Print the results DataFrame
print("\nSummary of Model Performance:")
print(results_df)
```
